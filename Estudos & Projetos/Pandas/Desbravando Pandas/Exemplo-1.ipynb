Neste exemplo, vamos demonstrar como ordenar uma coluna e remover duplicatas. Quando trabalhamos com um dataset que contém uma coluna de horas, é importante identificar a última atualização e eliminar as entradas duplicadas.



```python
df = pd.DataFrame(df)
```
