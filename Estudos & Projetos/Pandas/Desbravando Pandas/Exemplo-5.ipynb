```python
import pandas as pd

# Lendo o arquivo Excel
df = pd.read_excel("../Dataset/Dataset.xlsx")

# Exibindo as primeiras linhas do dataset para verificar
df.head()
```

```python
# Filtrando os dados para o usu√°rio 'dcta462j-313l-515fm-t6gh-6424er7845m2'
# e somando os pontos

condicao = df['IDCustomer']=='dcta462j-313l-515fm-t6gh-6424er7845m2'
df_user = df[condicao]
df_user['Points'].sum()
```

```python
# Loop para somar os pontos de todos os usu√°rios √∫nicos e armazenando em um dici√°rio

pontos = {}

for i in df['IDCustomer'].unique():
    condicao = df['IDCustomer']==i
    pontos[i] = df[condicao]['Points'].sum()

pontos

# Essa opera√ß√£o √© muito custosa, pois ela percorre o dataset inteiro para cada usu√°rio √∫nico ou seja n√£o √© uma boa pr√°tica
```

```python
# Utilizando o groupby para somar os pontos de todos os usu√°rios √∫nicos
# Essa √© a maneira mais inteligente de agrupar usu√°rios e somar os pontos

df_summary = df.groupby(['IDCustomer'])['Points'].sum()
df_summary.reset_index()
```

```python
# Utilizando o groupby para somar os pontos e contar o n√∫mero de UUID de todos os usu√°rios √∫nicos e a √∫ltima data de transa√ß√£o

df.groupby(['IDCustomer']).agg({'Points':'sum', 'UUID': 'count', 'DTTransaction': 'max'}).reset_index().rename(columns={'Points':'TotalPoints', 'UUID':'Frequencia', 'DTTransaction':'UltimaTransacao'})
```

```python
import datetime

# Calculando a diferen√ßa entre a √∫ltima transa√ß√£o e a data atual

data_1 = df['DTTransaction'][0]
now = datetime.datetime.now()

data = (now - data_1).days

data
```

```python
# Calculando quantos dias o usu√°rio est√° sem transacionar

condicao = df['IDCustomer']=='dcta838j-174l-151fm-t6gh-3006er7845m2'
df_user = df[condicao]
diff = datetime.datetime.now() - df_user['DTTransaction'].max()

diff.days
```

```python
# Fun√ß√£o para calcular a rec√™ncia e aplicando no groupby
def recencia(x):
    diff = datetime.datetime.now() - x.max()
    return diff.days


df.groupby(['IDCustomer']).agg({'Points':'sum', 'UUID': 'count', 'DTTransaction': ['max', recencia]}).reset_index().rename(columns={'Points':'TotalPoints', 'UUID':'Frequencia', 'DTTransaction':'UltimaTransacao'})
```
