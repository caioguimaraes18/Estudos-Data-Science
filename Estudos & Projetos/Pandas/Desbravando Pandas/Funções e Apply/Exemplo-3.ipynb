```python
import pandas as pd

# Lendo o arquivo Excel
df = pd.read_excel("../Dataset/Dataset.xlsx")

# Exibindo as primeiras linhas do dataset para verificar
df.head()
```

```python
df["Points_double"] = df["Points"] * 2
df
```

```python
df["Points_div"] = df["Points_double"] / df["Points"]
df
```

```python
df["Constante"] = 1
df
```

```python
# Criando uma nova coluna com base em uma função

get_f = lambda UUID: UUID.split("-")[0]
get_f("1a2b3c4d-5e6f-7g8h-9i0j")
```

```python
df["UUID"] = df["UUID"].apply(get_f)
df
```

```python
# Outra maneira de fazer a mesma coisa é utilizando o método applu com uma função lambda

df["UUID"] = df["UUID"].apply(lambda UUID: UUID.split("-")[0])
df
```

```python
def intervalo_pontos(pontos):
    if pontos < 100:
        return "Baixo"
    elif pontos < 500:
        return "Médio"
    else:
        return "Alto"
    
df["Intervalo_Pontos"] = df["Points"].apply(intervalo_pontos)
df
```
